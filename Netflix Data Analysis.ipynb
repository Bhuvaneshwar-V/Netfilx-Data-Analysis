# Q. 2) In which year the highest number of the TV Shows & Movies were released ? Show with Bar Graph

```python
df.dtypes # to chcek for how the date format is
```

```python
# changing object date datatype to datetime datatype by creating new column with the help of old column
df['Date_N'] = pd.to_datetime(df['Release_Date'])
```

```python
df.head()
```

```python
df.dtypes # checking the new column data type
```

```python
df.Date_N.dt.year.value_counts() # It counts the occurrence of all individual years in date column
# dt.year - considering year only
```

```python
# Bar Graph
df.Date_N.dt.year.value_counts().plot(kind='bar')
plt.xlabel("Year")
plt.ylabel("No. of. Movies")
```
